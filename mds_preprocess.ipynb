```python
import numpy as np
import pandas as pd

data = pd.read_csv("raw_nov30.csv")
data = data.drop(data[data['Finished'] == '0'].index)
data_col_name = ['MTR',	'Bus',	'Shuttle Bus',	'Taxi',	'Tram']
data = data.drop([0, 1], axis = 0)
mdsdata = data.loc[:, "Q11_1":"Q11_5"]
mdsdata.columns= data_col_name
```

```python
mdsdata = mdsdata.T
mdsdata
```

```python
def convertpref(i):
    if i == 1:
        return 5
    elif i == 2:
        return 4
    elif i == 3:
        return 3
    elif i == 4:
        return 2
    elif i == 5:
        return 1
```

```python
mdsdata = mdsdata.apply(pd.to_numeric)
mdsdata = mdsdata.applymap(convertpref)
mdsdata
```

```python
mdsdata.to_csv("MDS.csv")
```
